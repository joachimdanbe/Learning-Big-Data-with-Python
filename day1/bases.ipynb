Handled non-detection values ("Non-detect" and "ND") by converting them to NaN

```python
df['Cs-134'] = df['Cs-134'].apply(lambda x : np.nan if x =="Non-detect" else x)

df['Cs-134']
```

```python
df.loc[:, columns]
```

```python
df.loc[:, columns] = df.loc[:, columns].map(lambda x : np.nan if x == "Non-detect" else x)

df.loc[:,columns] = df.loc[:,columns].map(lambda x : np.nan if x == "ND" else x)
```

```python
df
```

```python
df.loc[:,['State', 'Location', 'Sample Type', 'Unit']] = df.loc[:,['State', 'Location', 'Sample Type', 'Unit']].map(lambda x : x.strip())
```

```python
df.dtypes
```

Converted date columns to proper datetime format

```python
df['Date Posted'] = pd.to_datetime(df['Date Posted'])
df['Date Collected'] = pd.to_datetime(df['Date Collected'])


#Converted all measurement columns to numeric type using pd.to_numeric()
for col in columns :
  df[col] = pd.to_numeric(df[col])
```

All types are okay

```python
df.dtypes
```

```python
mean = df['Cs-134'].mean()

df['Cs-134'] = df['Cs-134'].fillna(mean)

df['Cs-134']
```

```python
mean = df.loc[:,columns].mean()

df.loc[:,columns] = df.loc[:,columns].fillna(mean)
```

```python
df.head()
```

***Educational objective*** was : Classical preprocessing of a scientific dataset with handling of missing values, cleaning of texts and conversion of types - essential preliminary step before analysis.

```python
df['State'] = df['State'].astype('category')
df['Location'] = df['Location'].astype('category')
df['Sample Type'] = df['Sample Type'].astype('category')
df['Unit'] = df['Unit'].astype('category')
```

```python
df.dtypes
```

```python
#1 Intercating with the python shell using the Ipython commands

vec = np.random.randint(0,100, 10)

vec
```

sort vec using numpy

```python
vec_sorted = np.sort(vec)
print(vec_sorted)
```

using for loop

```python
for j in np.arange(vec.size):
  val = vec[j]
  i = j-1
  while i>=0 and vec[i]>val:
    vec[i+1] = vec[i]
    i = i-1
  vec[i+1] = val

print(vec)
```
